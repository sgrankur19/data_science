### Regression Model -
#### Q5. Split data into training and validation sets and build a simple Linear Regression model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_squared_log_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Time-based split: first 80% for train, last 20% for validation
split_idx = int(len(df) * 0.8)
train_df = df.iloc[:split_idx].copy()
val_df = df.iloc[split_idx:].copy()

print(f'Train: {len(train_df):,} rows, Val: {len(val_df):,} rows')
print(f'Train date range: {train_df.datetime.min()} to {train_df.datetime.max()}')
print(f'Val date range: {val_df.datetime.min()} to {val_df.datetime.max()}')

# Features (top informative + basics)
features = ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 
            'humidity', 'windspeed', 'hour', 'dayofweek', 'month']
X_train = train_df[features]
y_train = np.log1p(train_df['count'])  # Log transform target
X_val = val_df[features]
y_val = np.log1p(val_df['count'])

# Scale features (important for linear regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Train Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)

# Predictions
train_pred = np.expm1(lr_model.predict(X_train_scaled))  # Back transform
val_pred = np.expm1(lr_model.predict(X_val_scaled))
y_train_orig = np.expm1(y_train)
y_val_orig = np.expm1(y_val)
```

```python
# Evaluate
train_rmse = mean_squared_error(y_train_orig, train_pred)**0.5
val_rmse = mean_squared_error(y_val_orig, val_pred)**0.5
val_rmsle = mean_squared_log_error(y_val_orig, val_pred)**0.5

print('\n=== Linear Regression Results ===')
print('Train RMSE:', )
print('Validation RMSE: {val_rmse:.2f}', val_rmse)
print('Validation RMSLE: {val_rmsle:.4f}')
```

#### Q6. To improve model performance
##### Extend feature space using polynomial transformations (degree 2 or 3)
##### Apply Ridge and Lasso

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.metrics import mean_squared_error, mean_squared_log_error


# Top features only (hour, temp dominate)
features = ['hour', 'temp', 'dayofweek', 'humidity', 'month']
X_train = train_df[features]
y_train = np.log1p(train_df['count'])
X_val = val_df[features]
y_val = np.log1p(val_df['count'])

# Baseline Linear Regression RMSE
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
baseline = LinearRegression().fit(X_train_scaled, y_train)
baseline_pred = np.expm1(baseline.predict(X_val_scaled))
baseline_rmse = mean_squared_error(np.expm1(y_val), baseline_pred)**0.5
print(f'Baseline Linear RMSE: {baseline_rmse:.1f}')

# Polynomial + Regularization experiments
best_rmse = baseline_rmse
results = []

for degree in [2, 3]:
    print(f'\n=== Degree {degree} Polynomial ===')
    
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_train_poly = poly.fit_transform(X_train)
    X_val_poly = poly.transform(X_val)
    
    print(f'Features: {X_train_poly.shape[1]}')
    
    # Scale
    scaler_poly = StandardScaler()
    X_train_scaled = scaler_poly.fit_transform(X_train_poly)
    X_val_scaled = scaler_poly.transform(X_val_poly)
    
    # Ridge
    for alpha in [0.1, 1.0, 10.0, 100.0]:
        ridge = Ridge(alpha=alpha)
        ridge.fit(X_train_scaled, y_train)
        pred = np.expm1(ridge.predict(X_val_scaled))
        rmse = mean_squared_error(np.expm1(y_val), pred)**0.5
        print(f'Ridge alpha={alpha}: RMSE={rmse:.1f}')
        results.append({'degree': degree, 'model': 'Ridge', 'alpha': alpha, 'rmse': rmse})
        
        if rmse < best_rmse:
            best_rmse = rmse
    
    # Lasso
    for alpha in [0.01, 0.1, 1.0, 10.0]:
        lasso = Lasso(alpha=alpha, max_iter=5000)
        lasso.fit(X_train_scaled, y_train)
        pred = np.expm1(lasso.predict(X_val_scaled))
        rmse = mean_squared_error(np.expm1(y_val), pred)**0.5
        print(f'Lasso alpha={alpha}: RMSE={rmse:.1f}')
        results.append({'degree': degree, 'model': 'Lasso', 'alpha': alpha, 'rmse': rmse})
```

### Model Comparison and Interpretation 
#### Q7. Summarize all results (of different models tried out) in one table (RMSLE, key observations).

```python
# Results summary
results_df = pd.DataFrame(results).sort_values('rmse')
print(f'\n=== BEST MODELS ===')
print(results_df.head())
print(f'Improvement: {((baseline_rmse - results_df.rmse.iloc[0])/baseline_rmse*100):.1f}%')
```

#### Q8. Plot residuals for the best model.

```python
# Train Ridge (BEST model)
poly = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_val_poly = poly.transform(X_val)

scaler_poly = StandardScaler()
X_train_scaled = scaler_poly.fit_transform(X_train_poly)
X_val_scaled = scaler_poly.transform(X_val_poly)

ridge = Ridge(alpha=10.0)
ridge.fit(X_train_scaled, y_train)
r_pred = np.expm1(ridge.predict(X_val_scaled))

y_val_orig = np.expm1(y_val)
# Calculate residuals
residuals = y_val_orig - r_pred
val_df['predicted'] = r_pred
val_df['residuals'] = residuals
```

```python
plt.scatter(val_df['predicted'], val_df['residuals'], alpha=0.6, s=20)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

scaler_poly = StandardScaler()
X_train_scaled = scaler_poly.fit_transform(X_train_poly)
X_test_scaled = scaler_poly.transform(X_test_poly)

ridge = Ridge(alpha=10.0)
ridge.fit(X_train_scaled, y_train)
test_predictions = np.expm1(ridge.predict(X_test_scaled))
#test_predictions = np.expm1(test_predictions_log) 
```
