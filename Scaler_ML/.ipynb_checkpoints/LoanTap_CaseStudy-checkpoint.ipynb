Imputing Missing Values

```python
# Converting emp_length data type to numerical for missing value imputaion
import re
def emp_length_numerical(s):
  if pd.isna(s):
    return np.nan
  s = str(s).lower().strip()
  if "<" in s:
    return 0
  match = re.search(r"(\d+)",s)
  if match:
    return int(match.group(1))
  return np.nan

df["emp_length"] = df["emp_length"].apply(emp_length_numerical)
```

```python
# emp_title
df['emp_title'] = df['emp_title'].fillna('Missing_employment_detail')
# Emp_length
# df["emp_length"] = df["emp_length"].apply(emp_length_numerical)
df["emp_length"] = df["emp_length"].fillna(df["emp_length"].median())
# Title
df['title']=df['title'].fillna('Missing_title')
# revol_util
df['revol_util']=df['revol_util'].fillna(df['revol_util'].median())
# mort_acc
df['mort_acc']=df['mort_acc'].fillna(0)
# pub_rec_bankruptcies
df['pub_rec_bankruptcies']=df['pub_rec_bankruptcies'].fillna(0)
```

*  All numerical variables has outliers and we will proceed with outlier treatment

```python
for col in df_num:
    low = df[col].quantile(0.01)
    high = df[col].quantile(0.99)
    print(f"{col}: cap between {low:.2f} and {high:.2f}")
```

```python
# Before capping
df_num.describe()
```

```python
#Capping variables using 1st & 99th Percentile considering financial dataset. Because we don't want to loose any
# possible real world case which could impact our model performance.
for col in df_num:
    low = df[col].quantile(0.01)
    high = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=low, upper=high)
```

### Feature Engineering

```python
# Creating Binary flags for attributes
df_model['pub_rec_flag'] = (df_model['pub_rec']>0).astype(int)
df_model['mort_acc_flag'] = (df_model['mort_acc']>0).astype(int)
df_model['pub_rec_bankruptcies_flag'] = (df_model['pub_rec_bankruptcies']>0).astype(int)
df_model['High_revol_util'] = (df_model['revol_util'] >80).astype(int) # High credit use is a risk
df_model['term_flag'] = (df_model['term']==' 60 months').astype(int)
```

```python
# Time based Features
date_cols = ['issue_d','earliest_cr_line']
for col in date_cols:
  df_model[col] = pd.to_datetime(df_model[col])

df_model['issue_year'] = df_model['issue_d'].dt.year
df_model['issue_month'] = df_model['issue_d'].dt.month
df_model['earliest_cr_line_age'] = ((df_model['issue_d'] - df_model['earliest_cr_line']).dt.days/365).round(0)
```

```python
# Geographical Features
df_model['state'] = df_model['address'].str.extract(r',\s*([A-Z]{2})\s*\d{5}')
```

* Additional New atributes based on credit:
- revol_bal/anual_inc
- Loan_amnt/annual_inc
- Installment/annual_inc


```python
df_model['revol_bal_to_income'] = df['revol_bal']/(df['annual_inc'] + 1)
df_model['loan_to_income'] = df['loan_amnt']/(df['annual_inc']+1)
df_model['installment_to_income'] = df['installment']/(df['annual_inc']/12)

# +1 to handle division by 0 error, in case of 0 annual income
```

Encoding Categorical Variables

```python
#Binary Encoding [0,1] for Target Variable:
df['loan_status'] = df['loan_status'].map({'Fully Paid':0,'Charged Off':1})
```

```python
# df['loan_status']
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

x = df.drop(columns = ['loan_status'],axis = 1)
y = df['loan_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 100)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Ordinal Encoding, due to order of values
grade_order = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}
x_train['grade'] = x_train['grade'].map(grade_order)
x_test['grade']  =x_train['grade'].map(grade_order)

subgrade_order = {sg:i+1 for i,sg in enumerate(sorted(df['sub_grade'].unique()))}
x_train['sub_grade'] = x_train['sub_grade'].map(subgrade_order)
x_test['sub_grade'] = x_test['sub_grade'].map(subgrade_order)
```

```python
# One hot encoding, due to small number of categories
one_hot_cols = ['home_ownership', 'verification_status', 'purpose','initial_list_status', 'application_type']

combined = pd.concat([x_train, x_test], axis = 0)
combined = pd.get_dummies(combined, columns = one_hot_cols, drop_first = True)

x_train = combined.iloc[:len(x_train),:]
x_test = combined.iloc[len(x_train):,:]
```
