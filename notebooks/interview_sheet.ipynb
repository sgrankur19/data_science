```python
def return_unique(l):

 res= []

 for i in l:
    if i not in res:
        res.append(i)
 return res
l=(1,2,2,2,3,3,4,5,5,6,6)
return_unique(l)
```

```python
def sort_sequence(s):
  temp = []

  for i in sorted(s.split('-')):
    temp.append(i)

  return "-".join(temp)

s = 'green-red-yellow-black-white'
sort_sequence(s)
```

```python
def is_even(l):
    temp=[]

    for i in l:
        if i%2==0:
            temp.append(i)

    return (temp)

is_even([1,2,3,4,5,6,7,8,9])
```
