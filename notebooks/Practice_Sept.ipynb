```python
def unique(l):
    res=[]
    for i in l:
        if i not in res:
            res.append(i)

    return res
l=[1,2,3,4,5]
unique(l)
```

```python
def seq(l):
    temp=[]
    for i in sorted(l.split('-')):
        temp.append(i)

    return "-".join(temp)

l='green-red-yellow-black-white'

seq(l)
```

```python
def fun(l):
    uc=0
    lc=0

    for i in l:
        if i.islower():
            lc+=1
        elif i.isupper():
            uc+=1
        else:
            pass

    return lc,uc


l="CampusX is an Online Mwentorship Orogram fOr Enlightening Yopung mInds"
x,y=fun(l)

print("no of lc:", x)
print ("no of uc:", y)
```
